##### Feature Engineering
##### 1. Feature Selection


```python
# Since BMI_tags, Label, Unnammed:0 are highly corelated thus Those are dropped 
# Creating New dataframe for storing preprocessed data
df_pro = df.iloc[:,1:9]
df_pro
```

```python
# Unique Values in Activities
act = df_pro['activity_level'].unique()
act
```

```python
# Since activities has only 5 unique values thus we can classify them on a scale of 1 to 5 
df_pro.loc[df['activity_level']==1.2,'activity_level'] = 1.0
df_pro.loc[df['activity_level']==1.3,'activity_level'] = 2.0
df_pro.loc[df['activity_level']==1.5,'activity_level'] = 3.0
df_pro.loc[df['activity_level']==1.7,'activity_level'] = 4.0
df_pro.loc[df['activity_level']==1.9,'activity_level'] = 5.0
# let's check the changes which are made 
act = df_pro['activity_level'].unique()
act
```

### EDA on preprocessed data

```python
# correlation matrix of final selected features 
correlation_matrix = df_pro.corr()
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# Plot to show activity level of different age groups
sns.violinplot(data=df_pro, x='activity_level', y='age')
plt.title('Relationship between age and activity_level')
plt.xlabel('age')
plt.ylabel('activity_level')
plt.grid()
plt.show()
```

```python
# plot showing relation between gender and bmr
sns.violinplot(data=df_pro,x='gender',y='BMR')
plt.title('Relationship between gender and BMR')
plt.xlabel('Gender')
plt.ylabel('BMR')
plt.grid()
plt.show()
```

```python
# plot showing relation between gender and bmi
sns.violinplot(data=df_pro,x='gender',y='BMI')
plt.title('Relationship between gender and BMI')
plt.xlabel('Gender')
plt.ylabel('BMI')
plt.grid()
plt.show()
```
